### 两组数据的函数拟合

```python
import matplotlib.pyplot as plt
import numpy as np
import math
x = [4,7,10,30,50,100,300,500,1000,2000]
y = [0.3,0.25,0.2,0.1,0.08,0.03,0.01,0.005,0.003,0.001]
plt.plot(x,y)
plt.show()
#x,y曲线图
```

### 图一为x,y曲线图

```python
plt.plot(x,y,'ro')
plt.show()
#x,y散点图
```

### 图二为x,y散点图

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
 
#自定义函数 反比例形式
def func(x, a):
    return a/x

x = np.array(x)
y = np.array(y)
#非线性最小二乘法拟合
popt, pcov = curve_fit(func, x, y)
#获取popt里面是拟合系数
print(popt)
a = popt[0] 
#b = popt[1]
print('系数a:', a)
#print('系数b:', b)
#print('系数c:', c)
yvals = func(x,a) #拟合y值
wucha = 0
for i in range(0,10):
    wucha = wucha + abs((yvals[i] - y[i])/y[i])
print('系数pcov:', pcov)
print('系数yvals:', yvals)
print('误差为',wucha)
#绘图
plot1 = plt.plot(x, y, 's',label='original values') #原始值
plot2 = plt.plot(x, yvals, 'r',label='polyfit values') #拟合值
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=1) #指定legend的位置右上角
plt.title('curve_fit')
plt.show()
```

### 图三为用反比例函数的拟合结果：y = 1.4386 / x ,误差为4.085

```python
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
 
#自定义函数 反比例2形式
def func(x, a,b):
    return a/x+b

x = np.array(x)
y = np.array(y)
#非线性最小二乘法拟合
popt, pcov = curve_fit(func, x, y)
#获取popt里面是拟合系数
print(popt)
a = popt[0] 
b = popt[1]
print('系数a:', a)
print('系数b:', b)
wucha2 = 0
for i in range(0,10):
    wucha2 = wucha2 + abs((yvals[i] - y[i])/y[i])
#print('系数c:', c)
yvals = func(x,a,b) #拟合y值
print('系数pcov:', pcov)
print('系数yvals:', yvals)
print('误差为',wucha2)
#绘图
plot1 = plt.plot(x, y, 's',label='original values') #原始值
plot2 = plt.plot(x, yvals, 'r',label='polyfit values') #拟合值
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc=1) #指定legend的位置右上角
plt.title('curve_fit')
plt.show()
```
